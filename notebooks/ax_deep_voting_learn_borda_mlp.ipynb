# Supervised learning of a classical voting rule

```python
%cd ..

import exp_learn_borda

architecture = 'MLP'  # MLP, CNN, WEC, SetTransformer
election_sampling = {'probmodel': 'IC'}
random_seed = 42  # seed for reproducibility
batch_size = 32
learning_rate = 1e-4
```

### 100% Borda learned for 3 alternatives and 3 voters

```python
exp_learn_borda.experimentBorda(
    architecture=architecture,  # model architecture: word embedding classifier
    max_num_voters=3,  # maximum number of voters in a profile
    max_num_alternatives=3,  # maximum number of alternatives (candidates)
    election_sampling=election_sampling,  # uniform distribution over rankings
    num_gradient_steps=1000,  # number of gradient steps during training
    report_intervals=100,  # frequency of evaluation logging
    dev_dataset_size=100,  # size of the evaluation (dev) dataset
    random_seed=random_seed,  # seed for reproducibility
    batch_size=batch_size,  # size of each training batch
    learning_rate=learning_rate,  # learning rate for optimizer
    save_model=True,  # whether to save the trained model
    voting_rule='Borda',  # target voting rule for training labels
)
```
